```python
# What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?

df_higher_50k = df[df['salary'] == '>50K']

# advanced = (df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])).sum()

lista = ['Bachelors', 'Masters', 'Doctorate']

advanced = 0
for k, name in enumerate(lista):
    number = df_higher_50k[df_higher_50k['education'] == name]['education'].count()
    advanced += (df['education'] == name).sum()
    lista[k] = number

print(lista, sum(lista))
print(advanced)
print(f'{(sum(lista) / advanced) * 100:.2f}%')
```

```python
q1 = df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])
q2 = df['salary'] == '>50K'

higher_education_rich = round((q1 & q2).sum() / q1.sum() * 100, 1)
lower_education_rich = round((~q1 & q2).sum() / (~q1).sum() * 100, 1)

print(higher_education_rich, lower_education_rich)
print((q1 & q2).sum())
print(q1.sum())
```

```python
# What percentage of people without advanced education make more than 50K?

lista_without = ['HS-grad', '11th', '9th', 'Some-college',
       'Assoc-acdm', 'Assoc-voc', '7th-8th', 'Prof-school',
       '5th-6th', '10th', '1st-4th', 'Preschool', '12th']

lista_todos = []

for key, name in enumerate(lista_without):
    valor = df_higher_50k[df_higher_50k['education'] == name]['education'].count()
    valor_all = df[df['education'] == name]['education'].count()
    lista_without[key] = valor
    lista_todos.append(valor_all)

print(round((sum(lista_without) / sum(lista_todos)) * 100, 1))
```

```python
# What country has the highest percentage of people that earn >50K and what is that percentage?

porcentage = (df_higher_50k['native-country'].value_counts() / df['native-country'].value_counts()).max()

country = (df_higher_50k['native-country'].value_counts() / df['native-country'].value_counts()).sort_values(ascending=False).index[0]

print(f'{country} is the country with the highest percentage of people that earn >50K.')
print(f'{porcentage * 100:.1f}%')
```

```python
# Identify the most popular occupation for those who earn >50K in India.

most_popular_occupation = df[(df['native-country'] == 'India') & (df['salary'] == '>50K')]['occupation'].value_counts().index[0]

print(f'The most popular occupation for those who earn >50K in India is {most_popular_occupation}.')
```
